```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
ds=pd.read_csv('../datasets/GPA_data.csv')
ds.head()
```

```python
dummies = pd.get_dummies(ds[['Gender', 'Extra_Curricular']])
dummies.head()
```

```python
dummies.drop(['Gender_Female', 'Extra_Curricular_No'], axis=1, inplace=True)
dummies.head()
```

```python
ds_new = pd.concat([ds, dummies], axis=1)
ds_new
```

```python
ds_new.drop(['Gender', 'Extra_Curricular'], axis=1, inplace=True)
ds_new.head()
```

```python
model = LinearRegression()
```

```python
ds_new_vals = ds_new.values
data = np.concatenate([ds_new_vals[:, 1:5], ds_new_vals[:, 6:]], axis=1)
target = ds_new_vals[:, 5]

print(data.shape, target.shape)
```

```python
train_X, test_X, train_y, test_y = train_test_split(data, target, test_size=0.2)

print(train_X.shape, test_X.shape)
print(train_y.shape, test_y.shape)
```

```python
model.fit(train_X, train_y)
```

```python
model.score(train_X, train_y) * 100
```

```python
preds = model.predict(test_X)
mse = mean_squared_error(preds, test_y)
rmse = np.sqrt(mse)
```
